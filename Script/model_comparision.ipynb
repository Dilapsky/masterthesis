```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path = '../Data/results/OK_result3/'
```

```python
def Loss(df1,df2):
    total_loss = 0.0
    for item in df1.columns.tolist()[1:]:
        for i in range(144):
            total_loss += abs(df1.at[i,item]-df2.at[i,item])
    print('Total Loss in Average: ',total_loss)
```

```python
cnndf = pd.read_csv(os.path.join(path,'CNNresult.csv'))
```

```python
actualdf = pd.read_csv(os.path.join(path,'Actual.csv'))
```

```python
avgresultdf = pd.read_csv(os.path.join(path,'Avg_TTI.csv'))
```

```python
lstmdf = pd.read_csv(os.path.join(path,'lstm_base_new.csv'))
```

```python
lsdf = pd.read_csv(os.path.join(path,'lstm_base.csv'))
```

```python
lstmredf = pd.read_csv(os.path.join(path,'LSTM_result.csv'))
```

```python
Loss(cnndf,actualdf)#L1 Loss of results predicted by CNN
```

```python
Loss(avgresultdf,actualdf)#L1 Loss of results predicted by Historical Average
```

```python
Loss(lstmdf,actualdf)
```

```python
Loss(lsdf,actualdf)
```

```python
Loss(lstmredf,actualdf)
```
