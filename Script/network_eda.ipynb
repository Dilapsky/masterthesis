```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from shapely.geometry import MultiLineString
from shapely.wkt import dumps, loads
from folium import plugins
from scipy import stats
import folium
import json
```

```python
longtotal_total = []
latitotal_total = []
with open('../Data/Section_201810192021/road_boundary.txt','r',encoding = 'utf-8') as f:
    data = f.readlines()
    for i in range(len(data)-1):
        listdata = data[i+1].strip('\n').split('\t')
        MLS = loads(listdata[2])
        long_total = []
        lati_total = []
        for g in MLS.geoms:
            for c in g.coords:
                long_total = np.hstack((long_total,c[0]))
                lati_total = np.hstack((lati_total,c[1]))
        longtotal_total = np.hstack((longtotal_total,long_total))
        latitotal_total = np.hstack((latitotal_total,lati_total))
print(len(longtotal_total))
```

```python
plt.hist(x = longtotal_total, bins = 100, color = 'steelblue', edgecolor = 'black')
plt.xlabel('Longitude')
plt.ylabel('Frequency')
plt.title('Distribution of Longitute')
plt.show()
```

```python
plt.hist(x = latitotal_total, bins = 100, color = 'steelblue', edgecolor = 'black')
plt.xlabel('Latitude')
plt.ylabel('Frequency')
plt.title('Distribution of Latitude')
plt.show()
```

```python
plotdata = latitotal_total
print(np.mean(plotdata))
print(np.min(plotdata))
print(np.max(plotdata))
print(np.median(plotdata))
print(np.percentile(plotdata, [25, 50, 75]))
print(np.var(plotdata))
```
