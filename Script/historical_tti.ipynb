```python
import pandas as pd
from shapely.geometry import MultiLineString
from shapely.wkt import dumps, loads
from shapely.geometry import shape,mapping
from shapely.geometry import Polygon
import folium
import json
import numpy as np
import time,datetime
#totaltti = pd.read_csv('../Data/2018-10-01-00_2018-11-30-00_tti.csv',sep = ',',index_col = '1',usecols = [1,2,3,4,5])
totaltti = pd.read_csv('../Data/2018-10-01-00_2018-11-30-00_tti.csv',sep = ',',index_col ='Unnamed: 0')
totaltti.columns = ['obj_id', 'Time', 'TTI','Speed']
new_data = pd.read_csv('./road_length.csv', sep = '\t')
Full_Output =pd.DataFrame([],columns = ['Time','TTI'])
# range(3600)
for s in range(3602):
    Now_TimeStamp = 1538928000 + s*600
    timeArray = time.localtime(Now_TimeStamp)
    otherStyleTime = time.strftime("%Y-%m-%d %H:%M:%S",timeArray)
    mincsv = totaltti[totaltti['Time'] == otherStyleTime]
    print('Now process to: ',otherStyleTime)

    mergedf = pd.merge(new_data,mincsv,on = 'obj_id')
    nowtime = mergedf.iloc[0]['Time']
    mergedf = mergedf.fillna({'Time':nowtime,'TTI':int(0),'Speed':int(0)})
    nowlengthlist = np.array(mergedf['lengthlist'])
    nowtti = np.array(mergedf['TTI'])
    weight_tti = np.average(nowtti,weights = nowlengthlist)
    output_n = pd.DataFrame({'Time':nowtime,'TTI':weight_tti},index = [s])
    Full_Output = Full_Output.append(output_n,ignore_index = False)
  #  print(weight_tti)
Full_Output.to_csv('./Full_Output_4th.csv',index = 0)
```
